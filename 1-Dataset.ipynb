```python
import pandas as pd
import numpy as np
from sklearn.metrics import cohen_kappa_score
```

# Tagalog Kappa

```python
df1 = pd.read_csv('DE1Tl.csv')
df2 = pd.read_csv('DE2Tl.csv')
df3 = pd.read_csv('DE3Tl.csv')
```

```python
score=cohen_kappa_score(df1['Topics'],df2['Topics'],df3['Topics'])
print(score)
```

# Tagalog Data Exploration Drop null values

```python
df1 = df1.dropna(axis=0, subset=['Topics'])
df1
```

```python
df2 = df2.dropna(axis=0, subset=['Topics'])
df2
```

```python
df3 = df3.dropna(axis=0, subset=['Topics'])
df3
```

# Tagalog(Combine)

```python
df_combined = pd.DataFrame()
df_combined['tweets'] = df1['text']
df_combined['de1'] = df1['Topics']
df_combined['de2'] = df2['Topics']
df_combined['de3'] = df3['Topics']
df_combined
```

```python
df_combined['Topic'] = df_combined.mode(axis=1)[0]
df_combined
```

```python
df_combined = df_combined.drop(['de1','de2','de3'], axis=1)
df_combined
```

```python
df_combined.to_csv(r"Tagalog.csv", index = False)
```
